# Bus Dashboard Monitor

In this notebook, opencv feature detection will be applied to video file images of bus dashboard to extract bus dashboard display. You need an HDMI monitor to run this application.

References:

https://github.com/galenballew/SDC-Lane-and-Vehicle-Detection-Tracking/tree/master/Part%20I%20-%20Simple%20Lane%20Detection


### Step 1: Load the overlay

```python
from pynq.overlays.base import BaseOverlay
from pynq.lib.video import *

base = BaseOverlay("base.bit")
hdmi_out = base.video.hdmi_out
```

### Step 2: Initialize HDMI I/O

```python
# monitor configuration: 1920*1080 @ 60Hz
Mode = VideoMode(1920,1080,24)
hdmi_out = base.video.hdmi_out
hdmi_out.configure(Mode,PIXEL_BGR)
hdmi_out.start()
```

### Step 3: Load video from file

```python
import numpy as np
import cv2

# monitor (output) frame buffer size
frame_out_w = 1920
frame_out_h = 1080

# video file (input) configuration
frame_in_w = 1920
frame_in_h = 1080

videoIn = cv2.VideoCapture('/home/xilinx/jupyter_notebooks/Bus_Dashboard/HMDashboardVid.mp4')
videoIn.set(cv2.CAP_PROP_FRAME_WIDTH, frame_in_w);
videoIn.set(cv2.CAP_PROP_FRAME_HEIGHT, frame_in_h);
print("capture device is open: " + str(videoIn.isOpened()))
```

### Step 4: Capture frame from video file

```python
# Capture video file image
import numpy as np

ret, inframe = videoIn.read()

# Display video file image via HDMI Out
if (ret):      
    outframe = hdmi_out.newframe()
    outframe[0:1080,0:1920,:] = inframe[0:1080,0:1920,:]
    hdmi_out.writeframe(outframe)
else:
    raise RuntimeError("Failed to read from video file.")
```

### Step 5: Now use matplotlib to show image inside notebook

```python
# Output video file image as JPEG
%matplotlib inline 
from matplotlib import pyplot as plt
import numpy as np
plt.imshow(outframe[:,:,[2,1,0]])
plt.show()
```

### Step 6: Test feature extraction

```python
import cv2
import numpy as np

frame_1080p = np.zeros((1080,1920,3)).astype(np.uint8)
np_frame = outframe

# Converting BGR to HSV 
hsv = cv2.cvtColor(np_frame, cv2.COLOR_BGR2HSV) 
      
# Define range of red color in HSV 
lower_red = np.array([30,150,50]) 
upper_red = np.array([255,255,180]) 
      
# Create a red HSV colour boundary and threshold HSV image 
mask = cv2.inRange(hsv, lower_red, upper_red) 
  
# Bitwise-AND mask and original image 
res = cv2.bitwise_and(np_frame, np_frame, mask= mask) 
frame_edges = cv2.Canny(np_frame,100,200) 

# Save processed image to file
frame_1080p[0:1080,0:1920,0] = frame_edges[0:1080,0:1920]
frame_1080p[0:1080,0:1920,1] = frame_edges[0:1080,0:1920]
frame_1080p[0:1080,0:1920,2] = frame_edges[0:1080,0:1920]

import PIL.Image
img = PIL.Image.fromarray(frame_1080p)
img.save("/home/xilinx/jupyter_notebooks/Bus_Dashboard/processed.jpg")

img
```

### Step 7: Create function for processing incoming image

```python
def process_image(image):
    # Converting BGR to HSV 
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV) 
      
    # Define range of red color in HSV 
    lower_red = np.array([30,150,50]) 
    upper_red = np.array([255,255,180]) 
      
    # Create a red HSV colour boundary and threshold HSV image 
    mask = cv2.inRange(hsv, lower_red, upper_red) 
  
    # Bitwise-AND mask and original image 
    res = cv2.bitwise_and(image, image, mask= mask) 
    result = frame_edges = cv2.Canny(image,100,200) 
    return result
```

### Step 8: Test on Video

```python
import numpy as np
import cv2

frame_1080p = np.zeros((1080,1920,3)).astype(np.uint8)

# monitor (output) frame buffer size
frame_out_w = 1920
frame_out_h = 1080

# video file (input) configuration
frame_in_w = 1920
frame_in_h = 1080

videoIn = cv2.VideoCapture('/home/xilinx/jupyter_notebooks/Bus_Dashboard/HMDashboardVid.mp4')
videoIn.set(cv2.CAP_PROP_FRAME_WIDTH, frame_in_w);
videoIn.set(cv2.CAP_PROP_FRAME_HEIGHT, frame_in_h);

while(1):
    ret, inframe = videoIn.read()

    # Display video file image via HDMI Out
    if (ret):    
        processed = process_image(inframe)
        
        frame_1080p[0:1080,0:1920,0] = processed[0:1080,0:1920]
        frame_1080p[0:1080,0:1920,1] = processed[0:1080,0:1920]
        frame_1080p[0:1080,0:1920,2] = processed[0:1080,0:1920]

        outframe = hdmi_out.newframe()
        outframe[0:1080,0:1920,] = frame_1080p[0:1080,0:1920,]
        hdmi_out.writeframe(outframe)
    else:
        raise RuntimeError("Failed to read from video file.")
```

### Step 9: Release video  file and HDMI

```python
videoIn.release()
hdmi_out.stop()
del hdmi_out
```
